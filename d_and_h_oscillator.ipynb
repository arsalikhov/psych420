```python
import matplotlib.pyplot as plt
```

```python
def s_of_t(dt, v, s):
    return (s + (v * dt))

def v_of_t(dt, a, v):
    return (v + (a * dt))

def a_of_t(p, s): 
    return((-1 * p * s))

def a_of_t_damped(p, s, k, v):
    return ((- 1 * p * s)  - (k * v))
```

```python
def release_spring_harmonic(repeat_n = 5):
    """Function that maps the location of the end of a fristionless spring over time

    Args:
        repeat_n (int, optional): Number of iterations to go over. Defaults to 5.

    Returns:
        t [list]: list of times
        s [list]: list of locations
    """    
    v = []
    s = []
    init_s = 10
    init_v = 0
    t = []
    p = 2
    dt = 0.05


    for i in range(repeat_n):
        t.append(dt*i)
        if i == 0:
            a = a_of_t(p, init_s)
            v.append(v_of_t(dt, a, init_v))
            s.append(s_of_t(dt, init_v, init_s))
        else:
            a = a_of_t(p, s[i-1])
            v.append(v_of_t(dt, a, v[i-1]))
            s.append(s_of_t(dt, v[i], s[i-1]))
            # v = v_of_t(dt, a, v)

    return t, s

t, s = release_spring_harmonic(repeat_n = 1000)

plt.title("Frictioneless Spring")
plt.xlabel("Time (sec.)")
plt.ylabel("Location")
plt.plot(t,s)
plt.show
```

```python
def release_spring_damped(repeat_n = 5):
    """Function that maps the location of the end of a damped spring over time

    Args:
        repeat_n (int, optional): Number of iterations to go over. Defaults to 5.

    Returns:
        t [list]: list of times
        s [list]: list of locations
    """  

    v = []
    s = []
    init_s = 10
    init_v = 0
    t = []
    p = 2
    dt = 0.05
    k = 1


    for i in range(repeat_n):
        t.append(dt*i)
        if i == 0:
            a = a_of_t_damped(p, init_s, k, init_v)
            v.append(v_of_t(dt, a, init_v))
            s.append(s_of_t(dt, init_v, init_s))

        else:
            a = a_of_t_damped(p, s[i-1], k, v[i-1])
            v.append(v_of_t(dt, a, v[i-1]))
            s.append(s_of_t(dt, v[i], s[i-1]))
            # v = v_of_t(dt, a, v)

    return t, s

t, s = release_spring_damped(repeat_n = 1000)

plt.title("Damped Spring")
plt.xlabel("Time (sec.)")
plt.ylabel("Location")
plt.plot(t,s)
plt.show
```
